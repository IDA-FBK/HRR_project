```python
# Function to simplify age groups
def simplify_age_group(age_str):
    if not isinstance(age_str, str):
        return 'Unknown'
    
    # Extract the lower bound number
    try:
        if '+' in age_str:
            lower = int(age_str.replace('+', ''))
        elif '-' in age_str:
            lower = int(age_str.split('-')[0])
        elif '<' in age_str:
            return '<60' # Already formatted
        else:
            return age_str
            
        # Logic: If younger than 60, group them. Else keep original.
        if lower < 60:
            return '<60'
        else:
            return age_str
            
    except ValueError:
        return age_str

df['age_group_simplified'] = df['age_group'].apply(simplify_age_group)

age_stats = df.groupby('age_group_simplified')['is_readmitted'].agg(['mean', 'count']).reset_index()
age_stats.columns = ['age_group', 'readmission_rate', 'hospitalization_count']
global_rate = df['is_readmitted'].mean()

def get_sort_key(age_str):
    if age_str == '<60':
        return 0  # Forces this to be first
    if '+' in age_str:
        return int(age_str.replace('+', ''))
    try:
        return int(age_str.split('-')[0])
    except:
        return 999

age_stats['sort_key'] = age_stats['age_group'].apply(get_sort_key)
age_stats = age_stats.sort_values('sort_key')

fig, ax1 = plt.subplots(figsize=(20, 10))
sns.set_style('whitegrid')
plt.grid(visible=False)

# Bar chart (Readmission Rate)
bars = sns.barplot(
    x='age_group', 
    y='readmission_rate', 
    data=age_stats, 
    ax=ax1, 
    color='skyblue', 
    alpha=0.8
)

# Add Percentages
for p in bars.patches:
    height = p.get_height()
    if height > 0:
        ax1.text(
            p.get_x() + p.get_width() / 2.,
            height + 0.002,
            f'{height:.1%}', 
            ha='center', 
            va='bottom', 
            fontsize=12, 
            color='black'
        )

ax1.axhline(y=global_rate, color='grey', linestyle='--', linewidth=1.5, label='Global Average')
ax1.set_xlabel('Age Group', fontsize=12)
ax1.set_ylabel('Readmission Rate', fontsize=12, color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.set_xticklabels(age_stats['age_group'], rotation=45)

# Line chart (Volume)
ax2 = ax1.twinx()
ax2.grid(False)
sns.lineplot(
    x='age_group', 
    y='hospitalization_count', 
    data=age_stats, 
    ax=ax2, 
    color='orange', 
    marker='o', 
    linewidth=2.5, 
    label='Hospitalization Volume'
)
ax2.set_ylabel('Hospitalization Volume', fontsize=12, color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Legend
legend_elements = [
    mpatches.Patch(color='skyblue', label='Readmission Rate'),
    Line2D([0], [0], color='orange', marker='o', label='Hospitalization Volume'),
    Line2D([0], [0], color='grey', linestyle='--', label='Global Average')
]
plt.legend(handles=legend_elements, loc='upper left')

plt.title('Readmission Rate and Hospitalization Volume by Age Group', fontsize=18)
plt.tight_layout()
plt.savefig('../plots/readmission_rate/readmission_rate_hospitalization_volume_vs_age_group.png')
plt.show()
```

```python
def group_los(days):
    if days == 1: return '1'
    elif days == 2: return '2'
    elif days == 3: return '3'
    elif days <= 6: return '4-6'
    elif days <= 13: return '7-13'
    else: return '14+'

df['los_grouped'] = df['num_days'].apply(group_los)
# count episodes per LOS bucket
print(df['los_grouped'].value_counts())
# 2. Enforce the correct categorical order for the X-axis
los_order = ['1', '2', '3', '4-6','7-13','14+'] # lace index intervals

# Calculate statistics per LOS bucket
los_stats = df.groupby('los_grouped').agg(
    readmission_rate=('is_readmitted', 'mean'),
    volume=('episode_id', 'count')
).reset_index()

# Sort dataframe by our defined order, not alphabetically
los_stats['los_grouped'] = pd.Categorical(los_stats['los_grouped'], categories=los_order, ordered=True)
los_stats = los_stats.sort_values('los_grouped')

global_rate = df['is_readmitted'].mean()

fig, ax1 = plt.subplots(figsize=(20, 10))
sns.set_style('whitegrid')
plt.grid(visible=False)  

# Bar chart for Readmission Rate (Left Y-Axis)
bars = sns.barplot(
    x='los_grouped', 
    y='readmission_rate', 
    data=los_stats, 
    ax=ax1, 
    color='skyblue', 
    alpha=0.8
)

# Add Readmission % on top of bars (matches your Age/CCI charts)
for p in bars.patches:
    height = p.get_height()
    # Adding a slight offset so the text floats cleanly above the bar
    ax1.text(p.get_x() + p.get_width() / 2., height + 0.002, 
             f'{height:.1%}', ha='center', va='bottom', fontsize=11, color='black')

ax1.axhline(y=global_rate, color='grey', linestyle='--', linewidth=1.5)
ax1.set_ylabel('Readmission Rate', fontsize=12, color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.set_xlabel('Length of Stay (Days)', fontsize=12)

# Line chart for Hospitalization Volume (Right Y-Axis)
ax2 = ax1.twinx()
ax2.grid(False) 
sns.lineplot(
    x='los_grouped', 
    y='volume', 
    data=los_stats, 
    ax=ax2, 
    color='orange', 
    marker='o', 
    linewidth=2,
    sort=False # Keeps the categorical order intact
)
ax2.set_ylabel('Hospitalization Volume', color='orange', fontsize=12)
ax2.tick_params(axis='y', labelcolor='orange')

# Combined Legend
legend_elements = [
    mpatches.Patch(facecolor='skyblue', edgecolor='none', label='Readmission Rate'),
    Line2D([0], [0], color='orange', marker='o', label='Hospitalization Volume'),
    Line2D([0], [0], color='grey', linestyle='--', label='Global Average')  
]
# Placed in upper left so it doesn't overlap the long-tail data on the right
ax1.legend(handles=legend_elements, loc='lower right', fontsize=11)

plt.title('Readmission Rate and Hospitalization Volume vs Length of Stay', fontsize=18)
plt.tight_layout()

# Save the plot
plt.savefig('../plots/readmission_rate/readmission_rate_hospitalization_volume_vs_length_of_stay.png')
plt.show()
```

```python
# readmission rate vs num_comorbidities and hospitalization volume
# Create grouped column

def cci_risk_group(x):
    if x <= 1:
        return 'Low (0-1)'
    elif 2 <= x <= 3:
        return 'Moderate (2-3)'
    elif 4 <= x <= 5:
        return 'High (4-5)'
    elif 6 <= x <= 10:
        return 'Very High (6-10)'
    else:
        return 'Extreme (11+)'

df['cci_grouped'] = df['age_adj_comorbidity_score'].apply(cci_risk_group)

# Aggregation
cci_stats = df.groupby('cci_grouped').agg({
    'is_readmitted': 'mean',
    'episode_id': 'count'
}).reset_index()

cci_stats.columns = ['cci_grouped', 'readmission_rate', 'hospitalization_count']

# Sorting
sort_map = {
    'Low (0-1)': 0,
    'Moderate (2-3)': 1,
    'High (4-5)': 2,
    'Very High (6-10)': 3,
    'Extreme (11+)': 4
}

cci_stats['sort_key'] = cci_stats['cci_grouped'].map(sort_map)
cci_stats = cci_stats.sort_values('sort_key')

# Calculate Global Average Readmission Rate
global_rate = df['is_readmitted'].mean()

# Plotting
fig, ax1 = plt.subplots(figsize=(20, 10))
sns.set_style('whitegrid')
plt.grid(visible=False)  # Disable grid for a cleaner look
# Bar chart (Readmission Rate)
bars = sns.barplot(
    x='cci_grouped',
    y='readmission_rate',
    data=cci_stats,
    ax=ax1,
    color='skyblue',
    alpha=0.9
)

# Add Values on Bars
for p in bars.patches:
    height = p.get_height()
    ax1.text(
        p.get_x() + p.get_width() / 2.,
        height + 0.002,
        f'{height:.1%}', 
        ha='center', 
        va='bottom', 
        fontsize=10, 
        color='black'
    )

# Add Global Average Line
ax1.axhline(y=global_rate, color='grey', linestyle='--', linewidth=1.5)
ax1.set_xlabel('Charlson Comorbidity Index (CCI) Group', fontsize=12)
ax1.set_ylabel('Readmission Rate', fontsize=12, color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.set_title('Readmission Rate and Hospitalization Volume by Charlson Comorbidity Index (CCI)', fontsize=18)

# Line chart (Hospitalization Volume)
ax2 = ax1.twinx()
ax2.grid(False)  # Disable grid on the right y-axis
sns.lineplot(
    x='cci_grouped',
    y='hospitalization_count',
    data=cci_stats,
    ax=ax2,
    color='orange',
    marker='o',
    linewidth=2.5,
    label='Hospitalization Volume'
)
ax2.set_ylabel('Hospitalization Volume', fontsize=12, color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Legend
legend_elements = [
    mpatches.Patch(facecolor='skyblue', edgecolor='none', label='Readmission Rate'),
    Line2D([0], [0], color='orange', marker='o', label='Hospitalization Volume'),
    Line2D([0], [0], color='grey', linestyle='--', label='Global Average')
]
ax2.legend(handles=legend_elements, loc='upper left')

plt.tight_layout()
plt.savefig('../plots/readmission_rate/readmission_rate_hospitalization_volume_vs_comorbidities.png')
```

```python
def categorize_complexity(x):
    if x == 0:
        return '0'
    elif x == 1:
        return '1'
    elif x == 2:
        return '2'
    else:
        return '3+'

def regroup_age(age_str):
    clean_str = str(age_str).replace('+', '')
    lower_bound = int(clean_str.split('-')[0])
    if lower_bound < 44:
        return '<44'
    else:
        return age_str

df['Comorbidity Level'] = df['num_comorbidities'].apply(categorize_complexity)
df['age_group_grouped'] = df['age_group'].apply(regroup_age)

def get_sort_key(age_str):
    if age_str == '<44':
        return 0
    else:
        clean = age_str.replace('+', '').split('-')[0]
        return int(clean)

df['age_sort_key'] = df['age_group_grouped'].apply(get_sort_key)
df = df.sort_values('age_sort_key')

fig, ax = plt.subplots(figsize=(20, 10))
sns.set_style("whitegrid")

sns.boxenplot(
    x='age_group_grouped',
    y='num_medications_at_discharge',
    hue='Comorbidity Level',
    data=df,
    ax=ax,
    hue_order=['0', '1', '2', '3+'], # Explicit order
    palette="viridis" 
)

# Styling
ax.set_title('The Polypharmacy Burden: Medication Load by Age and Complexity', fontsize=18, pad=20)
ax.set_xlabel('Age Group', fontsize=12)
ax.set_ylabel('Number of Medications at Discharge', fontsize=12)
ax.tick_params(axis='x', rotation=0) # Rotation 0 usually works fine if groups are merged

# Legend
ax.legend(title='Comorbidities Count', loc='upper left', fontsize=11, title_fontsize=12)

plt.tight_layout()
plt.savefig('../plots/medications/medication_burden_by_age_grouped_and_comorbidity.png')
plt.show()
```

```python
def categorize_ambulatory(x):
    if x <= 5:
        return 'Low (0-5)'
    elif x <= 12:
        return 'Moderate (6-12)'
    elif x <= 24:
        return 'High (13-24)'
    else:
        return 'Extreme (25+)'

df['ambulatory_grouped'] = df['count_prior_ambulatory'].apply(categorize_ambulatory)

group_order = ['Low (0-5)', 'Moderate (6-12)', 'High (13-24)', 'Extreme (25+)']

stats = df.groupby('ambulatory_grouped').agg({
    'is_readmitted': 'mean',
    'episode_id': 'count' # Counting episodes = Hospitalization Volume
}).reindex(group_order).reset_index()

stats.columns = ['ambulatory_group', 'readmission_rate', 'hospitalization_volume']

global_rate = df['is_readmitted'].mean()

fig, ax1 = plt.subplots(figsize=(20, 10))
sns.set_style("whitegrid")
plt.grid(visible=False)

bars = sns.barplot(
    x='ambulatory_group',
    y='readmission_rate',
    data=stats,
    ax=ax1,
    color='skyblue',
    alpha=0.85
)

# Annotate Bars with Percentages
for p in bars.patches:
    height = p.get_height()
    if height > 0:
        ax1.text(
            p.get_x() + p.get_width() / 2.,
            height + 0.002,
            f'{height:.1%}', 
            ha='center', 
            va='bottom', 
            fontsize=12, 
            color='black',
        )

ax1.axhline(y=global_rate, color='grey', linestyle='--', linewidth=1.5, label='Global Average')

# Axis Styling (Left)
ax1.set_xlabel('Prior Ambulatory Engagement Level', fontsize=12)
ax1.set_ylabel('Readmission Rate', fontsize=12, color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.set_title('The Engagement Spectrum: Readmission Risk by Outpatient Intensity', fontsize=18)

# Line Chart: Hospitalization Volume (Right Axis)
ax2 = ax1.twinx()
ax2.grid(False)
# Scale the axis to give the line some headroom so it doesn't overlap the bars
ax2.set_ylim(0, stats['hospitalization_volume'].max() * 1.2)

sns.lineplot(
    x='ambulatory_group',
    y='hospitalization_volume',
    data=stats,
    ax=ax2,
    color='orange',
    marker='o',
    linewidth=3,
    markersize=9,
    sort=False
)

ax2.set_ylabel('Hospitalization Volume', fontsize=12, color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Legend
legend_elements = [
    mpatches.Patch(facecolor='skyblue', label='Readmission Rate'),
    Line2D([0], [0], color='orange', marker='o', label='Hospitalization Volume'),
    Line2D([0], [0], color='black', linestyle='--', label='Global Average')
]
ax2.legend(handles=legend_elements, loc='upper left')

plt.tight_layout()
plt.savefig('../plots/readmission_rate/readmission_rate_by_ambulatory_visits.png')
plt.show()
```

```python
import upsetplot
import os

# Find the file
plotting_path = os.path.join(os.path.dirname(upsetplot.__file__), 'plotting.py')
print(f"Patching: {plotting_path}")

# Read
with open(plotting_path, 'r') as f:
    content = f.read()

# Patch the 4 broken inplace lines
content = content.replace(
    'styles["linewidth"].fillna(1, inplace=True)',
    'styles["linewidth"] = styles["linewidth"].fillna(1)'
)
content = content.replace(
    'styles["facecolor"].fillna(self._facecolor, inplace=True)',
    'styles["facecolor"] = styles["facecolor"].fillna(self._facecolor)'
)
content = content.replace(
    'styles["edgecolor"].fillna(styles["facecolor"], inplace=True)',
    'styles["edgecolor"] = styles["edgecolor"].fillna(styles["facecolor"])'
)
content = content.replace(
    'styles["linestyle"].fillna("solid", inplace=True)',
    'styles["linestyle"] = styles["linestyle"].fillna("solid")'
)

# Write back
with open(plotting_path, 'w') as f:
    f.write(content)

print("Patch applied successfully")
```

```python
from upsetplot import UpSet, from_contents
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
df = pd.read_csv('../data/processed/modeling/inpatient_episodes_not_filtered.csv')
condition_cols = ['HF', 'IHD', 'COPD', 'Pneumonia', 'Stroke']

patient_conditions = (
    df.groupby('crypted_patient_id')[condition_cols]
    .max()
    .reset_index()
    .copy()
)

data = from_contents({
    c: set(patient_conditions.loc[patient_conditions[c]==1, 'crypted_patient_id'])
    for c in condition_cols
})

fig = plt.figure(figsize=(18, 7))

upset = UpSet(
    data,
    subset_size='count',
    min_subset_size=30,
    show_counts=False,              # we add counts manually below
    facecolor='skyblue',
    other_dots_color='#D3D1C7',
    totals_plot_elements=0,         # ← removes left set size bars entirely
    sort_by='cardinality',
    sort_categories_by='-cardinality',
)

axes = upset.plot(fig)

# Add counts on top of each vertical bar manually
ax_inter = axes['intersections']
for bar in ax_inter.patches:
    height = bar.get_height()
    if height > 0:
        ax_inter.text(
            bar.get_x() + bar.get_width() / 2,
            height + 8,
            f'{int(height):,}',
            ha='center', va='bottom',
            fontsize=10, fontweight='500',
            color='#3d3d3a'
        )

# Formatting
ax_inter.axis('off')   # ← removes everything: ticks, spine, labels

# Re-add only the title text
ax_inter.text(
    -0.06, 0.5, 'Intersection size',
    transform=ax_inter.transAxes,
    fontsize=13, rotation=90,
    va='center', ha='center',
    color='#3d3d3a'
)
# Larger category labels on matrix
for text in axes['matrix'].get_yticklabels():
    text.set_fontsize(13)

fig.suptitle('Patient condition overlap', fontsize=18)
plt.tight_layout()
plt.show()
```
