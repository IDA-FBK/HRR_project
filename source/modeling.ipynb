```python
# Stratified 80/20 train-test split based on patient crypted ID
import pandas as pd
from sklearn.model_selection import train_test_split

patients = pd.read_csv("../data/patients_readmissions_flag.csv")
episodes = pd.read_csv("../data/inpatient_episode_labeled.csv")

train_pat, test_pat = train_test_split(
    patients,
    test_size=0.2, #80/20 split
    random_state=42, 
    stratify=patients["readmission_flag"] # Stratify based on readmission flag
)

train_ids = set(train_pat["crypted_patient_id"]) # Set of patient IDs in training set
test_ids  = set(test_pat["crypted_patient_id"]) # Set of patient IDs in test set

train_episodes = episodes[episodes["crypted_patient_id"].isin(train_ids)].copy() # Training episodes
test_episodes  = episodes[episodes["crypted_patient_id"].isin(test_ids)].copy() # Test episodes

train_episodes.to_csv("../data/inpatient_episodes_train.csv", index=False)
test_episodes.to_csv("../data/inpatient_episodes_test.csv", index=False)
```
